```python
def preprocess(text):
    texts = []
#     for sentence in sent_tokenize(text):
#         preprocessed_text = ' '.join([t for t in m.lemmatize(sentence) if re.search('[\w\d]', t)])
#         texts.append(preprocessed_text)
    return ' '.join([t for t in m.lemmatize(text) if re.search('[\w\d]', t)])
#     return texts # list of strings

def get_bigrams(text):
    text = preprocess(text).split()
    bigrams = []

    for n, word in enumerate(text[:-1]):
        bigrams.append((word, text[n + 1]))
            
    return bigrams

def get_word2bigram(dataset_list): # get word2bigram
    word2bigram = {}
    texts = []
    for text in tqdm(dataset_list):
        sentences = sent_tokenize(text)
        texts += sentences 
    
    for text in tqdm(texts):
        bigrams = get_bigrams(text)
        
        for bigram in bigrams:
            for lemma in bigram :
                word2bigram[lemma] = word2bigram.get(lemma, set()) | {bigram}
    
    return word2bigram

def search_in_word2bigram(text, word2bigram): # search in word2bigram
    res = [v for k,v in word2bigram.items() if text in k]
    res = [el for ngram in res for el in ngram]
#     for k,v in word2bigram.items():
#         if text in k:
#             print(k, v)
    return len(res), res
```
